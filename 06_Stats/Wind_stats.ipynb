### Assign it to a variable called data and replace the first 3 columns by a proper datetime index. 

### Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
wind['Yr_Mo_Dy'].unique()
```

```python
def minus_year(x):
    year = x.year
    if x.year > 1989:
        year -= 100 
    return datetime.date(year, x.month, x.day) 

wind['Yr_Mo_Dy'] = wind['Yr_Mo_Dy'].apply(minus_year)
```

### Downsample the record to a yearly frequency for each location.

- **다운샘플링:** 소수 클래스의 크기와 일치하도록 다수 클래스에서 데이터를 제거하여 데이터 세드의 불균형을 해결하는 일반적인 데이터 처리 기술

```python
# 연간 빈도로 다운샘플링 → 연(annual) 단위로 그룹화

wind.groupby(wind.index.to_period('A')).mean()
```

```python
wind.groupby(wind.index.to_period('Y')).mean()
```

- **to_period():** 특정 주기(period)를 나타내는 데이터 타입으로 날짜를 변환할 때 사용
- 'Y', 'M', 'D', 'H'(시간, Hour)
