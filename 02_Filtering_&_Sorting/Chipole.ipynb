### How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype('float')
```

```python
chipo['item_price'].dtypes
```

```python
ten = chipo[chipo['item_price'] >= 10]
ten
```

### What is the price of each item?

```python
# 'item_name, quantity를 모두 고려해 중복된 행 제거
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity']) 
chipo_filtered
```

```python
item_per_price = chipo_filtered.loc[chipo_filtered['quantity'] == 1, :] 
item_per_price
```

```python
item_per_price = item_per_price.drop(columns=['order_id', 'quantity', 'choice_description'])
item_per_price.sort_values(by='item_price', ascending=False)
```

### How many times was a Veggie Salad Bowl ordered?

```python
chipo_salad = chipo.loc[chipo['item_name'] == 'Veggie Salad Bowl']
chipo_salad['item_name'].value_counts()
```

→ 데이터 프레임 전체에 value_counts()를 호출하면 오류가 발생하거나 원하는 결과가 나오지 않을 가능성이 높다. 왜냐하면 value_counts()는 **특정 컬럼에 대해 실행하는 함수**인데, chipo_salad 자체는 여전히 데이터프레임 형태로 유지되기 때문이다. 

### How many times did someone order more than one Canned Soda?

```python
chipo_soda = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
chipo_soda['item_name'].value_counts()
```
