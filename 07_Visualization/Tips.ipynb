### **sns.histplot() vs sns.displot()** 
- **histplot(): 히스토그램을 그리는 함수**
- **displot(): 히스토그램 + KDE를 포함한 분포 시각화** 

```python
total = sns.histplot(tips['total_bill'])
total.set(xlabel = 'value', ylabel = 'frequency', title='Total Bill')

plt.show()
```

```python
total = sns.displot(tips['total_bill'])
total.set(xlabel = 'value', ylabel = 'frequency', title='Total Bill')

# seaborn에서 그래프의 테두리를 제거하는 함수 
sns.despine()
```

- **sns.despine()**: 기본적으로 그래프의 오른쪽과 위쪽 테두리를 제거
→ seaborn에서 matplotlib의 plt.show()처럼 그래프를 출력하는 함수는 따로 존재하지 않지만, seaborn은 matplotlib 기반으로 plt.show()를 그대로 사용가능 

### Create a scatter plot presenting the relationship between total_bill and tip.

```python
plt.scatter(x = tips['total_bill'], y = tips['tip'], c='green', alpha=0.7)

plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship between Total Bill and Tip")

plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

### Create one image with the relationship of total_bill, tip and size.

```python
sns.pairplot(tips)
```

### Present the relationship between days and total_bill value.

```python
sns.stripplot(x=tips['day'], y=tips['total_bill'], data=tips, jitter=True)
```

- jitter: 데이터 포인트가 겹치는 것을 방지하게 위해 점들을 랜덤하게 흩뿌리는 기능

### Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex.

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = tips, jitter = True)
```

### Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch).

```python
sns.boxplot(data=tips, x=tips['day'], y=tips['total_bill'], hue=tips['time'])
```

- **hue의 역할**: 데이터 포인트를 특정 그룹별로 색상을 다르게 표시하는 역할

### Create two histograms of the tip value based of Dinner and Lunch. They must be side by side.

```python
sns.set(style='ticks')

g = sns.FacetGrid(tips, col = "time")
g.map(plt.hist, "tip");
```
