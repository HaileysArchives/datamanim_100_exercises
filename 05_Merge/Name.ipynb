```python
import pandas as pd
import numpy as np
```

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
df1 = pd.DataFrame(raw_data_1, columns=['subject_id', 'first_name', 'last_name'])
df2 = pd.DataFrame(raw_data_2, columns=['subject_id', 'first_name', 'last_name'])
df3 = pd.DataFrame(raw_data_3, columns=['subject_id', 'test_id'])

df3.head()
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

### Join the two dataframes along rows and assign all_data

```python
print(type(df1))
```

```python
# 행에 따라 조인 
all_data = pd.concat([df1, df2])
all_data
```

### Join the two dataframe along columns and assing to all_data_col

```python
# 열에 따라 조인
all_data_col = pd.concat([df1, df2], axis=1)
all_data_col
```

### Print data3

```python
df3
```

### Merge all_data and data3 along the subject_id value

- **concat()과 merge()는 Pandas 라이브러리에 속한 함수이기에 pd.concat / pd.merge 붙이기**

```python
pd.merge(all_data, df3, on='subject_id', how='inner')
```

### Merge only the data that has the same 'subject_id' on both data1 and data2

```python
pd.merge(df1, df2, on='subject_id', how='inner')
```

### Merge all values in data1 and data2, with matching records from both sides where available.

```python
# 가능한 경우 양쪽에서 일치하는 레코드를 사용해 데이터1과 데이터2의 모든 값 병합
pd.merge(df1, df2, on='subject_id', how='outer')
```

- **how='inner': 공통된 키값만 결합(교집합) - 두 데이터프레임의 겹치는 부분만 포함**
- **how='outer': 모든 키 값 포함(합집합), 결측치 채움 - 모든 데이터 포함**포함
