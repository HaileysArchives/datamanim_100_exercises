```python
import pandas as pd
import numpy as np
```

```python
cars1 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv')
cars2 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv')

print(cars1.head())
print(cars2.head())
```

```python
print(cars1.info())
print('-'*50)
print(cars2.info())
```

### Oops, it seems our first dataset has some unnamed blank columns, fix cars1

```python
del cars1.loc[:, 'Unnamed: 9': 'Unnamed: 13']
cars1.columns

# del은 변수나 컬럼을 삭제하는 python 기본 키워드이지만, .loc[]을 사용한 슬라이싱을
# 직접 삭제하는 데 사용할 수 없음
```

```python
del cars1['Unnamed: 9']
cars1.info()
```

```python
cars1 = cars1.loc[:, 'mpg':'car']
cars1.columns
```

### What is the number of observations in each dataset?

```python
print(cars1.shape)
print(cars2.shape)
```

### Join cars1 and cars2 into a single DataFrame called cars

```python
print(cars1.columns)
print('-'*50)
print(cars2.columns)
```

```python
cars = cars1.merge(cars2, on='mpg', how='outer')
cars
```

- **merge():** 기본적으로 두 데이터프레임의 **공통된 값**을 기준으로 결합
> 공통된 key(열)을 기준으로 행을 결합하기 때문에, 열이 그대로 유지되고 행만 합쳐지는 것

행 수가 예상보다 많아지는 이유는 how='outer'을 사용하면 두 데이터프레임의 모든 행을 유지하면서, 공통된 키(mpg)값이 중복될 경우 각 조합이 생성되기 때문이다

```python
cars = cars1.merge(cars2, on='mpg', how='outer').drop_duplicates()
cars
```

```python
cars = cars1.merge(cars2)
cars.shape
```

```python
print(type(cars1))
print(type(cars2))
```

```python
cars = cars1.append(cars2)
cars

# 최근 Pandas에서 append()가 더 이상 지원되지 않음! → concat() 사용
```

- merge(): 공통된 열(column)을 기준으로 병합 - **수평결합(왼쪽-오른쪽)**
- append(): 단순히 행(row)을 추가 → 인덱스 기준 - **수직결합(위-아래)**

```python
cars = pd.concat(cars1, cars2)
cars
```

```python
# concat() 함수는 병합할 여러 개의 데이터프레임을 리스트로 묶어야 한다. 
# → 리스트 형식으로 데이터를 받는 이유는 여러 개의 데이터프레임을 한 번에 결합할 수 있도록 설계되었음 

cars = pd.concat([cars1, cars2]) 
cars
```

```python
cars.shape
```

#### merge() 사용방법
##### 1) 객체 방식: cars1.merge(cars2, on='mpg', how='inner')
##### 2) 함수 방식: pd.merge(cars1, cars2, on='mpg', how='inner') 

- pd.merge(): 공통 열 기준으로 병합
- pd.concat(): 단순 연결 → 열 개수 달라도 자동으로 NaN 채움
