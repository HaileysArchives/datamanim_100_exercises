### Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.

```python
stud_alcoh['Mjob'] = stud_alcoh['Mjob'].apply(capitalizer)
stud_alcoh['Fjob'] = stud_alcoh['Fjob'].apply(capitalizer)
stud_alcoh.tail()
```

- **agg(): 집계함수**, 숫자 데이터나 특정 집계 연산 (예를 들어 sum, mean에 적합)
- **apply() 함수**:각 행(row) 또는 각 원소에 함수 적용 가능 

### Create a function called majority that returns a boolean value to a new colun called legal_drinker (Consider majority as older than 17 years old)

```python
stud_alcoh['legal_drinker'] = stud_alcoh[stud_alcoh['age'] > 17]

# stud_alcoh[stud_alcoh['age'] > 17]는 조건을 만족하는 행들만 포함하는 부분 데이터프레임
# legal_drinker 컬럼은 기존 데이터프레임의 모든 행을 포함해야 하지만, 현재 코드에서는 
# 일부 행만 포함됨 → 인덱스 문제 발생
```

```python
# create a function (return boolean value) 
def majority(x):
    if x > 17:
        return True
    else:
        return False

# create a new column called legal_drinker 
stud_alcoh['legal_drinker'] = stud_alcoh['age'].apply(majority)
stud_alcoh
```

### Multiply every number of the dataset by 10.

```python
stud_alcoh.info()
```

```python
# create a function 
def multi(x):
    if type(x) is int:
        return 10 * x
    return x
```

```python
stud_alcoh.applymap(multi).head()
```

- **apply(): Series(열) 또는 DataFrame(행 단위 적용 가능)**
- **applymap(): DataFrame(각 개별 원소)에 적용 가능**
