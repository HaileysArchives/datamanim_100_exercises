### Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
crime.info()
```

### Set the Year column as the index of the dataframe.

```python
crime = crime.set_index(keys='Year', drop=True)
crime.head()
```

### Delete the Total column

```python
crime = crime.drop(columns=['Total'], axis=1)
```

### Group the year by decades and sum the values
Pay attention to the Population column number, summing this column is a mistake 

```python
# 연도를 수십년 단위로 그룹화하고 값을 환산 
crime = crime.resample('10AS').sum()
```

```python
population = crime['Population'].resample('10AS').max()
```

```python
crime['Population'] = population

crime.index = crime.index.year
crime
```

- **resample():** Pandas에서 시계열 데이터를 특정 시간 간격으로 그룹화하여 집계하는 함수
    → 시간 단위별로 데이터를 재구성하고, 평균, 합계 등의 연산을 수행할 수 있도록 도와줌


> '10AS'의 의미: 10년 단위로 재샘플링
