### Change the frequency to monthly, sum the values and assign it to monthly.

```python
# 주기를 월별로 변경 및 값 합산 후 월별 할당 → resample()
month = df.resample('M').sum()
month.head()
```

```python
month
```

```python
month.replace(0, np.nan, inplace=True) # 0을 NaN으로 변환 
month
```

### You will notice that it filled the dataFrame with months that don't have any data with NaN. Let's drop these rows.

```python
month = month.dropna()
month
```

### Good, now we have the monthly data. Now change the frequency to year.

```python
year = month.resample('Y').sum()
year.head()
```

- **'Y': 연도별 리샘플링 (각 연도의 마지막 날(12월 31일)을 기준으로 그룹화**
- **'AS-JAN': 연초 기준 리샘플링 (각 연도의 첫 벌째 달(1월)을 기준으로 그룹화)**

```python
year = month.resample('AS-JAN').sum()
year.head()
```
