### Which continent drinks more beer on average?

```python
# mean()을 올바르게 호출하지 않음
# continent 컬럼(문자열)과 mean 메서드 자체를 비교하려다보니 타입 에러

alcohol[alcohol['continent'] > (alcohol['beer_servings'].mean)]
```

```python
alcohol.groupby(by='continent').head()
```

```python
alcohol.groupby('continent')['beer_servings'].mean()
```

### For each continent print the statistics for wine consumption.

```python
alcohol.groupby('continent')['wine_servings'].describe()
```

### Print the mean alcohol consumption per continent for every column.

```python
# numeric_only=True 옵션을 사용하면 숫자 데이터만 대상으로 연산 수행 가능
alcohol.groupby('continent').mean(numeric_only=True)
```

### Print the median alcohol consumption per continent for every column

```python
alcohol.groupby('continent').head(2)
```

```python
alcohol.groupby('continent').median(numeric_only=True)
```

### Print the mean, min and max values for spirit consumption. 
#### This time output a DataFrame!

```python
alcohol.info()
```

```python
# agg() 함수: Pandas에서 그룹화된 데이터를 대상으로 여러 가지 집계 함수를 적용

alcohol.groupby('continent')['spirit_servings'].agg(['min', 'max', 'mean'])
```
