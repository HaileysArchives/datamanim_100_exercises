### Set the values of the first 3 rows from alcohol as NaN.

```python
wine.iloc[:3, :] = np.nan
wine.head()
```

### Now set the value of the rows 3 and 4 of magnesium as NaN.

**wine['magnesium']은 DataFrame이 아니라 Series이기 때문에, 2차원 인덱싱이 불가 → 따라서, wine['magnesium'].iloc[2:4, :] 오류 발생**

```python
wine.iloc[2:4, 3] = np.nan
```

```python
wine.head()
```

### Fill the value of NaN with the number 10 in alcohol and 100 in magnesium.

```python
wine['alcohol'].replace(np.nan, 0, inplace=True)

wine.head()
```

```python
wine['magnesium'].replace(np.nan, 100, inplace=True)
wine.head()
```

```python
wine['alcohol'].fillna(10, inplace=True)
wine['magnesium'].fillna(100, inplace=True)
wine.head()
```

- **fillna(): NaN(결측값)을 원하는 값으로 채워주는 함수**

### Count the number of missing values.

```python
wine.isnull().sum()
```

### Create an array of 10 random numbers up until 10.

```python
# 10 이하의 10개의 임의의 숫자로 구성된 배열 생성
random = np.random.randint(1, 11, size=10)
random
```

### Use random numbers you generated as an index and assign NaN value to each of cell.

```python
# 생성된 임의의 숫자를 인덱스로 사용하고 각 셀에 NaN 값 할당
wine = wine.set_index(random)
wine.index = np.nan
wine.head()
```

```python
# 내가 인덱스로 넣은 값은 10개 밖에 안돼서 오류 발생
wine['alcohol'][random] = np.nan
wine.head(10)
```

### How many missing values do we have?

```python
wine.isnull().sum()
```

### Delete the rows that contains missing values.

```python
# del vs drop() 
wine = wine.dropna(axis=0, how='any')
wine.head()

# inplace=False(기본값): 원본 wine은 그대로 유지되고, 결측치가 제거된 새로운 DataFrame 반환
# inplace=True: 원본 wine 데이터프레임 자체가 수정됨! 
```

- **dropna(how='any'): 하나라도 NaN이 있으면 제거**
- **dropna(how='all'): 모두 NaN일 때만 제거** 

- **drop(): 원하는 행 또는 열을 직접 지정해서 제거**
- **dropna(): 결측치(NaN)가 있는 행 또는 열을 자동으로 제거**
