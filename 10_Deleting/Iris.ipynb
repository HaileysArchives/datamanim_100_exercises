```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from collections import Counter

%matplotlib inline
```

```python
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
iris.head()
```

```python
iris.shape
```

```python
iris.info()
```

### Create columns for the dataset.

```python
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris.head()
```

```python
iris.info()
```

### Is there any missing value in the dataframe?

```python
iris.isnull().sum()
```

### Let's set the values of the rows 10 to 29 of the column 'petal_length' to NaN.

```python
iris.iloc[10:29, 2:3] = np.nan
iris.head(30)
```

### Good, now let's substitute the NaN values to 1.0.

```python
iris.replace('NaN', 1.0, inplace=True)
iris.head(30)

# 'NaN'은 실제 NaN값이 아니라 문자열이라 변환되지 않음!
# 따라서, np.nan으로 변경해주어야 함!
```

```python
iris.replace(np.nan, 1.0, inplace=True)
iris.head(30)
```

```python
iris['petal_length'].fillna(1.0, inplace=True)
iris.head(30)
```

### Now let's delete the column class.

```python
del iris['class']
```

```python
iris.columns
```

### Set the first 3 rows as NaN.

```python
iris.iloc[:3, :] = np.nan
iris.head()
```

### Delete the rows that have NaN.

```python
iris = iris.dropna(how='any') # how='any': NaN이 하나라도 있으면 제거 (기본값)
iris.head()
```

- **dropna(): 데이터프레임에서 NaN 값을 포함한 행 또는 열을 삭제**

### Reset the index so it begins with 0 again.

```python
iris = iris.reset_index(drop=True)
iris.head()
```

#### **reset_index():** 데이터프레임의 인덱스를 초기화하고 새로운 숫자 인덱스 생성
- **drop=True를 설정하면 기존 인덱스를 삭제하고 새로운 인덱스만 유지**
